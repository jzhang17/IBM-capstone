```python
import os, json
import pandas as pd


path_to_json = 'r/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]


jsons_data = pd.DataFrame(columns=["filing_type", "filing_date", "filename", "company_cik", "doc_type", "filing_id", "company_name", "id"])


for index, js in enumerate(json_files):
    with open(os.path.join(path_to_json, js)) as json_file:
        json_text = json.load(json_file)

        
        filing_type = json_text['filing_type']
        filing_date = json_text['filing_date']
        filename = json_text['filename']
        company_cik = json_text['company_cik']
        doc_type = json_text['doc_type']
        filing_id = json_text['filing_id']
        company_name = json_text['company_name']                           
        id = json_text['id']                           
       
        
        jsons_data.loc[index] = [filing_type, filing_date, filename, company_cik, doc_type, filing_id, company_name, id]


print(jsons_data)
```

```python
from IPython.display import HTML
import base64
def create_download_link(jsons_data, title = "Download CSV file", filename = "data.csv"):  
    csv = jsons_data.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(jsons_data)
```
